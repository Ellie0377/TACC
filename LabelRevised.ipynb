```python
df = pd.read_parquet("Dataset.parquet")

recovery_intervals = [
    (10, "recover 10 min"),
    (30, "recover 30 min"),
    (60, "recover 60 min"),
    (90, "recover 90 min")
]

# 為了避免覆蓋掉原始攻擊標籤，我們先處理攻擊標記，再處理恢復標記
for attack_id, start, end in attacks:
    
    start_dt = pd.to_datetime(start)
    end_dt = pd.to_datetime(end)
    
    # 2. 標記恢復期
    # 注意：我們從「攻擊結束時間 (end_dt)」開始往後算
    for minutes, label_name in recovery_intervals:
        rec_start = end_dt
        rec_end = end_dt + pd.Timedelta(minutes=minutes)
        
        # 建立遮罩：僅標記「尚未被標記」的資料 (避免 recover 90 蓋掉 recover 10)
        # 或者你可以由大到小標記，這裡我們用邏輯判斷：
        # 如果該欄位還是 '0'，才填入恢復標籤
        mask_rec = (df['Timestamp'] > rec_start) & (df['Timestamp'] <= rec_end) & (df['Detailed_Label'] == '0')
        
        df.loc[mask_rec, 'Detailed_Label'] = f"{attack_id}_{label_name}"

print("標記完成！")
print(df['Detailed_Label'].value_counts())
```
